# Temperature
Temperature is a parameter applied to the softmax function during token sampling that controls the probability distribution over the vocabulary — a lower temperature (e.g., 0.0–0.3) sharpens the distribution, making the model greedily favor high-probability tokens, while a higher temperature (e.g., 0.8–1.5) flattens it, giving lower-probability tokens a greater chance of being selected.

The practical effect is that low temperatures produce more deterministic, focused, and repetitive output, whereas high temperatures yield more diverse, creative, and occasionally incoherent responses.

As a rule of thumb, use low temperature for tasks requiring precision and consistency (code generation, factual Q&A, structured extraction) and higher temperature for creative writing, brainstorming, or any task where variety and exploration matter more than reliability.

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Helper methods
def add_user_message(messages, text):
    user_message= { "role": "user", "content": text }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message= { "role": "assistant", "content": text }
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 100,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params['system'] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
context = []

add_user_message(context, "Generate a one sentence movie idea")
```

```python
answer = chat(context, None, 0.0)
answer
```

```python
answer = chat(context, None, 0.5)
answer
```

```python
answer = chat(context, None, 1.0)
answer
```
