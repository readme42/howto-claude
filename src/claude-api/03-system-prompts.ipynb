# System prompts

## Math Tutor Specialist

Should
- Initially, only give the students hints
- Patiently walk the student through a solution
- Show solutions for similar problems

Should not
- Immediately give a direct
- Tell the student to use a calculator

```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Helper methods
def add_user_message(messages, text):
    user_message= { "role": "user", "content": text }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message= { "role": "assistant", "content": text }
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 100,
        "messages": messages
    }

    if system:
        params['system'] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
context = []

add_assistant_message(context, "How do I solve 5x+3=2 for x?")

system = """
You are a patient math tutor.
Do not directly answer a student's question.
Guide them to a solution step by step.
"""

answer = chat(context, system)
```
